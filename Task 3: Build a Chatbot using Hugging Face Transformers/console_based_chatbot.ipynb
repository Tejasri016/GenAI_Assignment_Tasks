```python
#import libraries
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

```

```python
#Load dataset
model_name = "microsoft/DialoGPT-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
chat_history_ids = None

print("Chatbot: Hello! I am your AI assistant. How can I help you today?")

while True:
    user_input = input("You: ")
    # exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye! Have a nice day 😊")
        break

    # Predefined responses for accuracy
    if user_input.lower() == "hello":
        print("Chatbot: Hello! Nice to meet you. How can I assist you today?")
        continue

    elif "artificial intelligence" in user_input.lower():
        print("Chatbot: Artificial Intelligence refers to the simulation of human intelligence by machines that can perform tasks such as learning, reasoning, and problem solving.")
        continue

    elif "who created python" in user_input.lower():
        print("Chatbot: Python was created by Guido van Rossum and released in 1991.")
        continue

    elif "thank you" in user_input.lower():
        print("Chatbot: You're welcome! Feel free to ask more questions.")
        continue

    # Transformer response
    new_input_ids = tokenizer.encode(
        user_input + tokenizer.eos_token,
        return_tensors='pt'
    )
    # append chat history
    bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids

    #generate response
    chat_history_ids = model.generate(
        bot_input_ids,
        max_length=200,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_k=60,
        top_p=0.95,
        temperature=0.9,
        no_repeat_ngram_size=2
    )

    #limit memory(for speed)
    chat_history_ids = chat_history_ids[:, -500:]

    #decode response
    response = tokenizer.decode(
        chat_history_ids[:, bot_input_ids.shape[-1]:][0],
        skip_special_tokens=True
    )

    print("Chatbot:", response)
```

**What is Transformer?**

A transformer is a deep learning model that processes text using attention mechanisms, allowing it to understand context efficiently.

**Why DialoGPT?**

DialoGPT is a conversational model designed for generating human-like dialogue responses.

**Limitation**

The chatbot may generate incorrect or generic responses because it is trained on conversational data rather than factual knowledge.
